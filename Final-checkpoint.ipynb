## Data Understanding

```python
#Lecture du fichier csv
url="NTL-DS-Phase-1-Project-Template/Data/Aviation_Data.csv"
df=pd.read_csv(url,low_memory=False)

#Apercu du fichier
df.dtypes
df.info()
```

## Data Preparation

#### Nettoyage des donnees

```python
#
print(f"Lignes : {df.shape[0]}, Colonnes : {df.shape[1]}")

#Clean columns with 
df_clean=df.dropna(axis=1, thresh=0.7*len(df))

#Informations sur les donnees recuperees
df_clean.info()

# Examen des donnees supprimees 
df_clean.isnull().sum().sort_values(ascending=False).head(10)
```

## Analysis and Results

```python
# Exemple : nombre d'accidents par type d'aÃ©ronef
df_clean['Engine.Type'].value_counts().plot(kind='bar', title="Nombre dâ€™accidents par catÃ©gorie d'engin")
plt.xlabel("Type d'engin")
plt.ylabel("Nombre d'accidents")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()


# Nombre total d'accidents
total = df.shape[0]
print(f"Nombre total d'accidents : {total}")

# RÃ©partition par type de vol
df["Purpose.of.flight"].value_counts().plot(kind="bar", title="Accidents par type de vol")
plt.ylabel("Nombre d'accidents")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

weather_counts = df["Weather.Condition"].value_counts()
weather_counts.plot(kind="pie", autopct="%1.1f%%", title="Conditions mÃ©tÃ©o lors des accidents", ylabel="")
plt.tight_layout()
plt.show()

df["Broad.phase.of.flight"].value_counts().plot(kind="barh", title="Accidents par phase de vol")
plt.xlabel("Nombre d'accidents")
plt.tight_layout()
plt.show()
```
